```python
import starsim as ss
import tbsim as mtb 
import matplotlib.pyplot as plt
```

```python
def make_tb_simplified(agents=1000, start=2000, dt=7/365):
    pop = ss.People(n_agents=agents)
    tb = mtb.TB(dict(beta = 0.001, init_prev = 0.25))
    net = ss.RandomNet(dict(n_contacts=ss.poisson(lam=5), dur = 0))
    dems = [ss.Pregnancy(pars=dict(fertility_rate=15)), ss.Deaths(pars=dict(death_rate=10))]
    sim = ss.Sim(people=pop, networks=net, diseases=tb, pars=dict(dt = dt, start = start), demographics=dems)
    sim.pars.verbose = sim.pars.dt / 5
    return sim
```

```python
sim_tb = make_tb_simplified(agents=1500, start=2000, dt=7/365)
sim_tb.run()
sim_tb.diseases['tb'].plot()
plt.show()
```

```python
def make_tb_Net(agents=1000, start=2000, dt=7/365, nt=None):
    pop = ss.People(n_agents=agents)
    tb = mtb.TB(dict(beta = 0.001, init_prev = 0.25))
    net = nt
    dems = [ss.Pregnancy(pars=dict(fertility_rate=15)), ss.Deaths(pars=dict(death_rate=10))]
    sim = ss.Sim(people=pop, networks=net, diseases=tb, pars=dict(dt = dt, start = start ), demographics=dems)
    sim.pars.verbose = sim.pars.dt / 5
    return sim
```

```python

sim_tb = make_tb_Net(agents=1000, 
                            start=1990, 
                            dt=7/365, 
                            nt=ss.RandomNet(dict(n_contacts=8))
)
sim_tb.run()
sim_tb.diseases['tb'].plot()
plt.show()
```

```python
sim_tb = make_tb_Net(agents=1000, 
                            start=1990, 
                            dt=.5,                                      # Half year
                            nt=[ss.MaternalNet(), ss.ErdosRenyiNet()]
)
sim_tb.run()
sim_tb.diseases['tb'].plot()
plt.show()
```

```python
sim_tb = None
sim_tb = make_tb_Net(agents=1000, 
                            start=1990, 
                            dt=.5,                    # Half year
                            nt=[ss.ErdosRenyiNet(dur = 0)]
)
sim_tb.run()
sim_tb.diseases['tb'].plot()
plt.show()
```

```python
sim_tb = make_tb_Net(agents=1000, 
                            start=1990, 
                            dt=12/365,
                            nt=[ss.RandomNet()]
)
sim_tb.run()
sim_tb.diseases['tb'].plot()
plt.show()
```
